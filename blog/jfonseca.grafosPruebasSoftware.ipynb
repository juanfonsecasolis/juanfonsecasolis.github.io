````python
import matplotlib.pyplot as plt
import networkx as nx # pip install networkx

# Generating sample data
#G = nx.florentine_families_graph()
G = nx.DiGraph()
G.add_edges_from([
    ('Creada', 'Decrementada'), ('Creada', 'Inventario corto'),
    ('Decrementada', 'Consolidada'),('Decrementada', 'Cancelada'), ('Decrementada', 'Consolidada'),
    ('Consolidada', 'Inventario corto'),('Consolidada', 'Cancelada'),('Consolidada', 'Enrutada'),
    ('Enrutada', 'Cancelada'),('Enrutada', 'Inventario corto'),
    ('Inventario corto', 'Decrementada'),('Inventario corto', 'Cancelada'), ('Inventario corto', 'Decrementada'),
    ('Enrutada', 'Impresa'),
    ('Impresa', 'Empacada'),
    ('Empacada', 'Despachada'),
    ('Despachada', 'Decrementada'),
    ('Decrementada', 'Finalizada')
    ])
#adjacency_matrix = nx.adjacency_matrix(G)

# The actual work
# You may prefer `nx.from_numpy_matrix`.
#G2 = nx.from_scipy_sparse_matrix(adjacency_matrix)
pos = nx.spring_layout(G, k=.25, iterations=10)
nx.draw(G,pos,with_labels=True,node_size = 1000)
plt.axis('equal')
````
